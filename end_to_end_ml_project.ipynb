**build train_test_split function for understanding..**

````python
import numpy as np

def train_test_split(data,test_ratio):
    np.random.seed(42)
    shuffled=np.random.permutation(len(data))
    #print(shuffled)                                # here every time we getting different set of values ..
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[test_indices],data.iloc[train_indices]

test_set,train_set=train_test_split(housing,0.2)


#print(f"test set length:{len(test_set)}\n train_set length={len(train_set)}")
````

**using better evalutin technique - cross validation**

Negative MSE: The scores returned by cross_val_score when using scoring="neg_mean_squared_error" are negative because scikit-learn aims to maximize all scoring metrics, even for error metrics. Therefore, it uses the negative MSE to make it compatible with maximization goals.

Positive MSE/RMSE: You negate the scores or take the square root to interpret them as positive error values, which makes them more intuitive. Lower RMSE or MSE means the model is performing better.

**All about cross-validation**


######Common Types of Cross-Validation
k-Fold Cross-Validation:

The dataset is split into k equal-sized subsets (folds).
The model is trained on k-1 folds and tested on the remaining one.
This process is repeated k times, with a different fold serving as the test set each time.
The average performance across all k iterations is reported as the final model performance.

Pros:

More reliable than a single train-test split.
Ensures each data point is used for both training and testing.
Cons:

Can be computationally expensive, especially for large datasets

## Decision tree regresstion**

When the number of features in the dataset is large, the process of dividing the tree (splitting) in **Decision Tree Regression** becomes more complex because the model has to consider splits along multiple feature dimensions rather than just one. Let's explore this with an example involving **multiple features**.

### Example: Dataset with Multiple Features

Suppose you have a dataset with two features (`X1`, `X2`) and a target variable (`Y`). The dataset looks like this:

| **X1** | **X2** | **Y** |
|--------|--------|-------|
| 2      | 3      | 5     |
| 3      | 6      | 9     |
| 4      | 8      | 11    |
| 5      | 2      | 7     |
| 6      | 4      | 10    |
| 7      | 6      | 14    |
| 8      | 9      | 17    |

Now, we have two features (`X1` and `X2`) that influence the target (`Y`). The Decision Tree Regressor must choose the best feature and the best split point for that feature at each node. 

---

### Step-by-Step Explanation of the Split

1. **Initial Data:**
   - There are 7 data points.
   - The goal is to split the data in a way that reduces the **Mean Squared Error (MSE)** the most.

#### Total Mean Squared Error (MSE) for the Dataset:

The mean of the target values:

\[
\bar{Y} = \frac{5 + 9 + 11 + 7 + 10 + 14 + 17}{7} = 10.43
\]

Now, calculate the initial **MSE** for the entire dataset:

\[
MSE = \frac{1}{7} \left( (5 - 10.43)^2 + (9 - 10.43)^2 + (11 - 10.43)^2 + (7 - 10.43)^2 + (10 - 10.43)^2 + (14 - 10.43)^2 + (17 - 10.43)^2 \right)
\]

\[
MSE = \frac{1}{7} \left( 29.49 + 2.04 + 0.33 + 11.76 + 0.19 + 12.72 + 43.17 \right)
\]

\[
MSE = \frac{1}{7} \times 99.7 = 14.24
\]

---

### Step 2: Evaluate Possible Splits for Each Feature

Now, the Decision Tree will consider splits along both `X1` and `X2`.

#### Split on `X1`:

Lets evaluate different possible splits for the feature `X1`:

1. **Split at `X1 = 3.5`:**

   - Group 1: \( X1 \leq 3.5 \)  Data points: \( (2, 3, 5), (3, 6, 9) \)
   - Group 2: \( X1 > 3.5 \)  Data points: \( (4, 8, 11), (5, 2, 7), (6, 4, 10), (7, 6, 14), (8, 9, 17) \)

   **MSE for Group 1**:

   The mean of `Y` in Group 1:

   \[
   \bar{Y}_{1} = \frac{5 + 9}{2} = 7
   \]

   MSE for Group 1:

   \[
   MSE_1 = \frac{1}{2} \left( (5 - 7)^2 + (9 - 7)^2 \right)
   \]

   \[
   MSE_1 = \frac{1}{2} \times \left( 4 + 4 \right) = 4
   \]

   **MSE for Group 2**:

   The mean of `Y` in Group 2:

   \[
   \bar{Y}_{2} = \frac{11 + 7 + 10 + 14 + 17}{5} = 11.8
   \]

   MSE for Group 2:

   \[
   MSE_2 = \frac{1}{5} \left( (11 - 11.8)^2 + (7 - 11.8)^2 + (10 - 11.8)^2 + (14 - 11.8)^2 + (17 - 11.8)^2 \right)
   \]

   \[
   MSE_2 = \frac{1}{5} \times \left( 0.64 + 23.04 + 3.24 + 4.84 + 26.44 \right) = \frac{1}{5} \times 58.2 = 11.64
   \]

   **Total MSE after the split at `X1 = 3.5`:**

   \[
   MSE_{split} = \frac{2}{7} \times 4 + \frac{5}{7} \times 11.64 = 1.14 + 8.31 = 9.45
   \]

---

#### Split on `X2`:

Now, lets evaluate different splits for the feature `X2`.

1. **Split at `X2 = 5`:**

   - Group 1: \( X2 \leq 5 \)  Data points: \( (2, 3, 5), (5, 2, 7), (6, 4, 10) \)
   - Group 2: \( X2 > 5 \)  Data points: \( (3, 6, 9), (4, 8, 11), (7, 6, 14), (8, 9, 17) \)

   **MSE for Group 1**:

   The mean of `Y` in Group 1:

   \[
   \bar{Y}_{1} = \frac{5 + 7 + 10}{3} = 7.33
   \]

   MSE for Group 1:

   \[
   MSE_1 = \frac{1}{3} \left( (5 - 7.33)^2 + (7 - 7.33)^2 + (10 - 7.33)^2 \right)
   \]

   \[
   MSE_1 = \frac{1}{3} \times \left( 5.43 + 0.11 + 7.11 \right) = \frac{1}{3} \times 12.65 = 4.22
   \]

   **MSE for Group 2**:

   The mean of `Y` in Group 2:

   \[
   \bar{Y}_{2} = \frac{9 + 11 + 14 + 17}{4} = 12.75
   \]

   MSE for Group 2:

   \[
   MSE_2 = \frac{1}{4} \left( (9 - 12.75)^2 + (11 - 12.75)^2 + (14 - 12.75)^2 + (17 - 12.75)^2 \right)
   \]

   \[
   MSE_2 = \frac{1}{4} \times \left( 14.06 + 3.06 + 1.56 + 17.06 \right) = \frac{1}{4} \times 35.74 = 8.94
   \]

   **Total MSE after the split at `X2 = 5`:**

   \[
   MSE_{split} = \frac{3}{7} \times 4.22 + \frac{4}{7} \times 8.94 = 1.81 + 5.11 = 6.92
   \]

---

### Step 3: Selecting the Best Split

Now we compare the MSE after each split:

- **Split at `X1 = 3.5`:** MSE = 9.45
- **Split at `X2 = 5`:** MSE = 6.92

The split at `X2 = 5` produces the lowest MSE, so the Decision Tree will choose to split on `X2` first.

### Step 4: Further Splitting

After the first split, the Decision Tree continues to split each group further. For instance, the left group \( X2 \leq 5 \) could be split again on either `X1` or `X2`, depending on which split minimizes the MSE for that group.

---

### Summary of the Process

1. **At each node**, the Decision Tree evaluates all possible splits for all features.
2. For each split, it calculates the **Mean Squared Error (MSE)** and selects the split that minimizes the weighted average MSE.
3. This process continues recursively until the stopping criteria are met (like a minimum number of samples per leaf or a maximum tree depth).
4. The final prediction in each leaf node is the **mean** of the target values in that node.

In a dataset with more features, the same process is repeated for each feature and each possible split point within that feature. The key point is that the tree finds splits
